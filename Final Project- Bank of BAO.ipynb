```python
import random
import json

# Create a method that could restore every account it loads:
##dictionary to store account details:


##This code snippet attempts to open and read a file named "accounts.json" in the current directory in read mode using a with statement. 
##The contents of the file are then loaded into a Python dictionary called accounts using the json.load() method.
try:
    with open('accounts.json', 'r') as f:
        accounts = json.load(f)
except FileNotFoundError:
    accounts = {}
 
 
#By using this statement, the file is automatically closed once the code inside the block is executed
##ensuring that the file is properly handled and any resources associated with it are released.#

def save_accounts():
    with open('accounts.json', 'w') as f:
        json.dump(accounts, f)
        
#Overall, this code snippet is used to load data from a JSON file into a Python dictionary to be used in further processing.##
        



##Function one: Account Creation##
def create_account():
    while True:    
        account_name = input("Enter your Name:")
        initial_deposit = input("Enter your initial deposit amount:")
        
        try:
            initial_deposit = float(initial_deposit)
        except ValueError:
            print("****Invalid input, please enter a numeric value.****")
            return 

        account_ID = random.randint(100000,999999)
        accounts[account_ID] = {"name": account_name, "balance": initial_deposit}   
        save_accounts()
                
        return f"Account created successfully! Hi, {account_name}, we have created a account for you, your account ID is: {account_ID}, your current balance is: {initial_deposit}"



##Function two: Deposit money##
def deposit_money():
    while True:
        account_ID = input("Enter your 6 digit account number:")
        if not account_ID.isnumeric():
            print("****Invalid account number, please enter a numeric value.****")
            continue
        if int(account_ID) not in accounts:
            print("****Invalid account number, please try again****")
            continue

        while True:
            deposit_amount = input("Enter your deposit amount:")
            try:
                deposit_amount = float(deposit_amount)
            except ValueError:
                print("****Invalid input, please enter a numeric value.****")
                continue

            accounts[int(account_ID)]["balance"] += deposit_amount
            save_accounts()

            return f"Deposited successfully! Your current balance is:{accounts[int(account_ID)]['balance']}"



        
        
##Function three: Withdraw Money##
def withdraw_money():
    while True:
        account_ID = input("Enter your 6 digit account number:")
        if not account_ID.isnumeric():
            print("****Invalid account number, please enter a numeric value.****")
            continue
        if int(account_ID) not in accounts:
            print("****Invalid account number, please try again****")
            continue


        while True:
            withdraw_amount = input("Enter your withdraw amount:")
            try:
                withdraw_amount = float(withdraw_amount)
            except ValueError:
                print("****Invalid input, please enter a numeric value.****")    
                continue
    
            if withdraw_amount > accounts[int(account_ID)]['balance']:
                print("****Insufficient balance, please input an amount****")   
                continue

            accounts[int(account_ID)]['balance'] -= withdraw_amount
            return f"In process, please remember to take away the cash. Your current balance is: {accounts[int(account_ID)]['balance']}"
      
    


        
##Function Four: Check Balance##      

def checkDeposit():
    while True:
        account_ID = input("Enter your 6 digit account number:")
        if not account_ID.isnumeric():
            print("****Invalid account number, please enter a numeric value.****")
            continue
        if int(account_ID) not in accounts:
            print("****Invalid account number, please try again****")
            continue
            
        while True:
            return f"Hi, your current balance is: {accounts[int(account_ID)]['balance']} "
        
        
        
        
        
        
while True:
    print("please select what you need today: \n\nPress 1 for creating a new customer\nPress 2 to deposit money\nPress 3 to withdraw money\nPress 4 to check balance\nPress exit to Exit" )
    service = input("Please select the service you want:")
    if service.lower() == 'exit':
        print("Goodbye!")
        break    
        
    if service.isdigit():
        service = int(service)

    else:
        print("\n****Invalid input, please enter a numeric value.****\n")       
        continue
        
    if int(service) == 1:
        print(create_account())
    
    elif int(service) == 2:
        print(deposit_money())
        
    elif int(service) == 3:
        print(withdraw_money())
        
    elif int(service) == 4:
        print(checkDeposit())        


    else:
        print("****Invalid service selected.****")
        

      
    
    print("--------------------------------------------")
```
